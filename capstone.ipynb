## Data Collection
# The dataset used in this project is the Zillow Home Value Index (ZHVI), obtained from Kaggle. 
#The data was stored in the raw data directory and loaded into the notebook using pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/raw/zhvi.csv")
df.head()
```

The dataset contains 313 observations and 52 columns, representing home values across different U.S. states over time. 

The "Unnamed: 0" column appears to represent dates and should be converted to a datetime format. Most variables are numeric, representing home values, with some missing values in certain states such as North Dakota and Montana.

## Missing Values and Data Quality

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

Some states contain missing values, with North Dakota and Montana having the most missing data. Most other states have complete data. These missing values may affect analysis and may need to be handled through removal or imputation.

## Exploratory Visualizations

```python
# Convert date column
df['Date'] = pd.to_datetime(df['Unnamed: 0'])
df.set_index('Date', inplace=True)
df.drop(columns=['Unnamed: 0'], inplace=True)

# Plot a few key states
df[['Florida', 'California', 'Texas', 'New York']].plot(figsize=(12,6))
plt.title("Home Value Trends Over Time")
plt.show()
```

## Key Insights

Insight 1:
Home values have increased steadily over time across major states such as California, Florida, and Texas.

Why it matters:
This suggests strong long-term growth in housing markets, making real estate an important financial asset for investment.

Insight 2:
There are significant differences in home values between states, with California and New York having much higher values than others.

Why it matters:
Location plays a critical role in housing prices, which is important for investors and policy decisions.

Insight 3:
Some states show more volatility in home prices than others.

Why it matters:
Understanding volatility helps assess financial risk when investing in different housing markets.

## Feature Distributions

```python
df[['Florida', 'California', 'Texas', 'New York']].hist(figsize=(10,8), bins=20)

plt.suptitle("Distribution of Home Values by State", fontsize=16)

# Add common labels
for ax in plt.gcf().axes:
    ax.set_xlabel("Home Value ($)")
    ax.set_ylabel("Frequency")

plt.show()
```

The distributions of home values show that some states have higher average values and wider spreads. States like California tend to have higher values, while others are more concentrated at lower ranges.

## Correlation Analysis

```python
corr = df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The correlation heatmap shows strong positive relationships between many states, suggesting that housing markets tend to move together. This indicates broader macroeconomic factors influencing home values.

## Feature Relationships

```python
sns.scatterplot(x=df['California'], y=df['Florida'])
plt.xlabel("California Home Values")
plt.ylabel("Florida Home Values")
plt.title("California vs Florida Home Values")
plt.show()
```

There is a strong positive relationship between home values in California and Florida, suggesting that housing markets across major states are interconnected.

```python
sns.lmplot(x='California', y='Florida', data=df.reset_index())
```

Insight:
Housing markets across different states are highly correlated, indicating that home values tend to move together over time.

Why it matters:
This suggests that housing is influenced by broad economic factors such as interest rates and inflation, rather than purely local conditions. For investors, this means that diversifying real estate investments across states may not significantly reduce risk, as markets tend to rise and fall together.

Insight:
There are significant differences in home value levels between states, with some states consistently having higher prices.

Why it matters:
This indicates that location plays a major role in housing affordability and investment potential, which is important for both buyers and policymakers.

```python
sns.scatterplot(x=df['Texas'], y=df['Florida'])
plt.title("Texas vs Florida Home Values")
plt.show()

sns.scatterplot(x=df['New York'], y=df['California'])
plt.title("New York vs California Home Values")
plt.show()
```
