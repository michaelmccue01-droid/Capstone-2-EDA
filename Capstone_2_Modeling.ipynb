Capstone Two: Modeling

Goal:
Predict future Florida Zillow Home Values using historical home value data and compare multiple regression models to identify the best forecasting approach.

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
X_train = pd.read_csv("data/X_train_scaled.csv", index_col=0)
X_test = pd.read_csv("data/X_test_scaled.csv", index_col=0)

y_train = pd.read_csv("data/y_train.csv", index_col=0).squeeze()
y_test = pd.read_csv("data/y_test.csv", index_col=0).squeeze()
```

The target variable is Florida home value.

Three regression models will be compared:

1. Linear Regression
2. Random Forest Regressor
3. Tuned Random Forest Regressor

```python
lr = LinearRegression()

lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
lr_mae = mean_absolute_error(y_test, y_pred_lr)

lr_rmse = np.sqrt(
    mean_squared_error(y_test, y_pred_lr)
)

lr_r2 = r2_score(y_test, y_pred_lr)

print("MAE:", lr_mae)
print("RMSE:", lr_rmse)
print("R2:", lr_r2)
```

```python
rf = RandomForestRegressor(
    random_state=42
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
rf_mae = mean_absolute_error(y_test, y_pred_rf)

rf_rmse = np.sqrt(
    mean_squared_error(y_test, y_pred_rf)
)

rf_r2 = r2_score(y_test, y_pred_rf)

print("MAE:", rf_mae)
print("RMSE:", rf_rmse)
print("R2:", rf_r2)
```

```python
param_grid = {
    'n_estimators':[50,100,200],
    'max_depth':[3,5,10,None]
}
```

```python
grid = GridSearchCV(
    RandomForestRegressor(
        random_state=42
    ),
    param_grid,
    cv=5
)

grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
best_rf = grid.best_estimator_

y_pred_best = best_rf.predict(X_test)
```

```python
best_mae = mean_absolute_error(
    y_test,
    y_pred_best
)

best_rmse = np.sqrt(
    mean_squared_error(
        y_test,
        y_pred_best
    )
)

best_r2 = r2_score(
    y_test,
    y_pred_best
)

print("MAE:", best_mae)
print("RMSE:", best_rmse)
print("R2:", best_r2)
```

```python
results = pd.DataFrame({
    'Model':[
        'Linear Regression',
        'Random Forest',
        'Tuned Random Forest'
    ],
    'MAE':[
        lr_mae,
        rf_mae,
        best_mae
    ],
    'RMSE':[
        lr_rmse,
        rf_rmse,
        best_rmse
    ],
    'R2':[
        lr_r2,
        rf_r2,
        best_r2
    ]
})

results
```

Conclusion

Three regression models were evaluated using MAE, RMSE, and R².

The model with the highest R² and lowest prediction error was selected as the final model.

Because this project involves forecasting a continuous variable (Florida home values), regression metrics are more appropriate than classification metrics such as accuracy or F1 score.

The selected model provides the best balance between predictive performance and practical implementation.
